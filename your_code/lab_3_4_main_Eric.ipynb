```python
def aggregate(x):
    return [stats.median(x), min(x), max(x), stats.stdev(x)]
    
employee.pivot_table(index = ['Department', 'Title'], values = ['Salary'], aggfunc = {'Salary':aggregate})
```

```python
# We end up with an error : due to the small size of the data set, some department/title pairs only have one employee
# This prevent from calculating a standard deviation
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
# your answer here
# If there is only one element in a dataset, the only value is automatically the mean and the median.
# The gap with the mean is thus 0 and we can replace the standard deviation by 0 when there is one and only employee.
```

```python
# New function
def agg_stdev(x):
    if len(x) == 1:
        return 0
    else:
        return round(stats.stdev(x),1)

employee.pivot_table(index = ['Department', 'Title'], values = ['Salary'], aggfunc = (stats.median, min, max, agg_stdev))
```

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# your answer here

def mean_without_max(x):
    if len(x) == 1:      # Here I exclude  the case where we have one employee only (happened when we looked at data per title)
        return 0
    else:
        max_index = x[x == max(x)].index[0]     # Find index of the maxima in the series, code inspired from https://stackoverflow.com/questions/18327624/find-elements-index-in-pandas-series
        y = x.drop(max_index, inplace = False)   # create a sublist without the maxima (not sure this step is necessary)
        return round(np.mean(y),1)       # round the result for better redability

employee.pivot_table(index = ['Department'], values = ['Salary'], aggfunc = {'Salary':mean_without_max})
```
